```python
ventas_totales_producto = df_sales.groupby(['ID_Tienda','Producto'])['Cantidad_Vendida'].sum().reset_index()
print(ventas_totales_producto)
```

```python
df_inventario_analisis = pd.merge(ventas_totales_producto, df_inventories, on=['ID_Tienda','Producto'])
df_inventario_analisis['Rotacion_Inventario'] = df_inventario_analisis['Cantidad_Vendida'] / df_inventario_analisis['Stock_Disponible']
tiendas_inventario_critico = df_inventario_analisis[df_inventario_analisis['Rotacion_Inventario'] < 0.1 ]
print(tiendas_inventario_critico)
```

```python
array_ventas = ventas_totales['Cantidad_Vendida'].to_numpy()
```

```python
mediana_ventas = np.median(array_ventas)
print(mediana_ventas)
```

```python
desviacion_ventas = np.std(array_ventas)
print(desviacion_ventas)
```

```python
np.random.seed(42)
mean_ventas = np.mean(array_ventas)
proyeccion = np.random.normal(loc=mean_ventas, scale=desviacion_ventas, size=10)
proyeccion = np.maximum(proyeccion, 0)
print(proyeccion)
```
